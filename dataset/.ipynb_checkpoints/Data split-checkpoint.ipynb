```python
import pickle
import pandas as pd
import json
import os
import numpy as np
from tqdm import tqdm
```

```python
data_path = 'train/pose_label.pkl'
with open(data_path, 'rb') as f:
    train = pickle.load(f, encoding='iso-8859-1')
```

```python
train_X = {}
for key, val in tqdm(train.items()):
    count_flag = 0
    for x, y in val:
        if not (x > 0 and y > 0):
            count_flag += 1
    if count_flag < 3: # 최소 16개의 keypoint가 있는 Pose
        train_X[key] = val # (x, y) -> (h, w)
```

```python
train_X
```

```python
length = len(train_X)
total_index = np.arange(length)
np.random.shuffle(total_index)
```

```python
train_ratio, test_ratio, valid_ratio = 0.7, 0.15, 0.15
train_num = int(length*0.7)
test_num = train_num + int(length*0.15)
valid_num = length # 끝까지
```

```python
train_index = total_index[:train_num]
test_index = total_index[train_num:test_num]
valid_index = total_index[test_num:]
```

```python
train, test, valid = {}, {}, {}
filenames = list(train_X.keys())

for i in train_index :
    filename = filenames[i]
    coord = train_X[filename]
    train[filename] = coord
    
for i in test_index :
    filename = filenames[i]
    coord = train_X[filename]
    test[filename] = coord
    
for i in valid_index :
    filename = filenames[i]
    coord = train_X[filename]
    valid[filename] = coord
```

```python
file_path = 'train.pkl'
file = train

with open(file_path, 'wb') as handle:
    pickle.dump(file, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open(file_path, 'rb') as handle:
    b = pickle.load(handle)

print(file == b)
```

```python
file_path = 'test.pkl'
file = test

with open(file_path, 'wb') as handle:
    pickle.dump(file, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open(file_path, 'rb') as handle:
    b = pickle.load(handle)

print(file == b)
```

```python
file_path = 'valid.pkl'
file = valid

with open(file_path, 'wb') as handle:
    pickle.dump(file, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open(file_path, 'rb') as handle:
    b = pickle.load(handle)

print(file == b)
```
